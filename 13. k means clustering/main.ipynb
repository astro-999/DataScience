```python
import pandas as pd 
df = pd.read_csv("https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv")
df
```

```python
x = df.iloc[:, 3:5].values #.values converts the dataframe into a numpy array
x 

```

```python
from  sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
#scatter plot of the data
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 6))
#unscaled scatter plot of the data
plt.scatter(x[:, 0], x[:, 1])
plt.title("Unscaled data")


```

```python
#scaled scatter plot of the data
fig = plt.figure(figsize=(10, 6))
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c = 'red')
plt.title("Scaled data")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.grid('True')
```

```python
#k-means clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster']= kmeans.fit_predict(x_scaled)
print(df['Cluster'])
#print the cluster centers
print(kmeans.cluster_centers_)

```

```python
#visualize the clusters
plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'cyan', 'magenta']

for i in range(5):
    plt.scatter(x[df['Cluster'] == i, 0], x[df['Cluster'] == i, 1], s = 50, c = colors[i], label = 'Cluster ' + str(i+1))

#plot the cluster centers
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], s = 200, c = 'yellow', label = 'Centroids', marker='*' ,edgecolors='black')

plt.title("K-means Clustering (scaled data)")
plt.xlabel("Annual Income (scaled)")
plt.ylabel("Spending Score (scaled)")
plt.legend()
```
