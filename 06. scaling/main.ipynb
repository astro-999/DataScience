```python
import pandas as pd
from sklearn.preprocessing import StandardScaler,MinMaxScaler, RobustScaler
prices = pd.Series([1, 2, 3, 4, 5], name='price')
prices

```

```python
data_2d = prices.values.reshape(-1, 1) # reshape the data to be 2D
data_2d
```

```python
#initialize the scaler 
scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()

#create a comparison dataframe 
comparison_df = pd.DataFrame({
    'original': prices,
    'standard_scaled': scaler.fit_transform(data_2d).flatten(), #flatten the data to be 1D
    'Normalized': minmax_scaler.fit_transform(data_2d).flatten(),#fit and transform the data
    'Robbust(IQR)': robust_scaler.fit_transform(data_2d).flatten()
})

comparison_df.round(2)
#fit and transform the data 
```
