```python
from scipy import stats
import numpy as np

#groupA : Sales without discount 
groupA =[200,210,190,205,195]

#groupB : Sales with discount
groupB =[250, 240,260,230,255]

# Perform t-test
t_statistic, p_value = stats.ttest_ind(groupA, groupB)
print(f"T-statistic: {t_statistic : .4f}")
print(f"P-value: {p_value : .4f}")

if p_value < 0.05:
    print("the discount has a significant effect on sales.")
    
else:  
    print("the discount does not have a significant effect on sales.")
```
