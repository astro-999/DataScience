```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Square Footage': [1500, 2000, 1200,2500, 1800, 1600, 2200,1400, 1900, 2100, 2800,1300,1700, 2400, 1100],
    'Bedroom': [3, 4, 2, 4, 3, 3,4,2, 3,4,5,2,3,4,2],
    'House_age':[10,5, 20 ,2, 15,8 ,1,12,7,4, 1,25,11,3,30],
    'price': [300,450, 200, 550, 350, 320, 480, 280, 380,420, 600, 180, 340, 500, 150 ]
})
df.head()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
X = df[['Square Footage', 'Bedroom', 'House_age']]
y = df['price']
model = LinearRegression().fit(X, y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

```

```python
predicted_house_price = model.predict([[2000, 3, 10]])
print("Predicted House Price for 2000 sqft, 3 bedrooms, and 10 years old:", predicted_house_price[0])

```
