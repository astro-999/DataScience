```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
```

```python
df = sns.load_dataset('titanic')
df
```

```python
features  = ['pclass','age','sex','fare','sibsp','parch']
x = df[features]
y = df['survived']
```

```python
from sklearn.impute import SimpleImputer # to handle missing values
from sklearn.pipeline import Pipeline # to create a pipeline for data preprocessing and modeling
from sklearn.compose import ColumnTransformer # to apply different transformations to different columns
from sklearn.preprocessing import OneHotEncoder # to handle categorical variables
# imputer = SimpleImputer(strategy='mean')
# x['age'] = imputer.fit_transform(x[['age']])
# x['fare'] = imputer.fit_transform(x[['fare']])
# x['sex'] = x['sex'].map({'male':0,'female':1})
```

```python

numeric_features = ['age','fare','sibsp','parch'] # these are the numeric features that we want to impute with median strategy
categorical_features = ['pclass','sex'] # these are the categorical features that we want to impute with most frequent strategy and then one hot encode them
numeric_transformer = SimpleImputer(strategy='median') # this is the transformer for numeric features that will impute missing values with median strategy
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder())
    
]) # this is the transformer for categorical features that will impute missing values with most frequent strategy and then one hot encode them
```

```python
Preprocessor = ColumnTransformer(transformers=[
    ('num',numeric_transformer,numeric_features),
    ('cat',categorical_transformer,categorical_features) 
]
)

Preprocessor
```

```python
from sklearn.model_selection import train_test_split
# splitting the data into training and testing sets with 80% for training and 20% for testing, and setting random state to 42 for reproducibility
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print(f'Training set shape: {x_train.shape}, Testing set shape: {x_test.shape}')
```

```python
model= Pipeline(steps=[
    ('preprocessor',Preprocessor),
    ('classifier',RandomForestClassifier(n_estimators=100,random_state=42))
])

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(cm,display_labels= ['Not survived ', 'survived']).plot(cmap = 'Blues')

print(cm)
```

```python

#F1 score is the harmonic mean of precision and recall, which gives a better measure of the model's performance when there is an imbalance in the classes. It ranges from 0 to 1, where 1 indicates perfect precision and recall, and 0 indicates poor performance.
f1 = f1_score(y_test,y_pred)
print(f'F1 Score: {f1:.2f}')
```
