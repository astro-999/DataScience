```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'X':[1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700],
    'Y':[245,312,279,308,199,219,405,324,319,255]
})
```

```python
sns .scatterplot(x='X',y='Y',data=df)
```

```python
model = LinearRegression().fit(df[['X']], df[['Y']])
print(f" coefficient : {model.coef_[0][0]:.2f}") 
print(model.intercept_)
print(model.coef_)
```

```python
#predicting the price of a house with 2000 sqft area
predicted_price = model.predict([[2000]])
print(predicted_price)

```

```python
#prediction of all y from all the x values
predicted_price_all = model.predict(df[['X']])
print(predicted_price_all)
#plotting the regression line
sns .scatterplot(x='X',y='Y',data=df)
plt.plot(df['X'],predicted_price_all,color='red')   

```

```python
sns .scatterplot(x='X',y='Y',data=df)
sns.lineplot(x=df['X'],y=predicted_price_all.flatten(),color='red')
```

```python
from sklearn.metrics import r2_score
r2= r2_score(df['Y'], predicted_price_all)
print(f"R-squared: {r2:.2f}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error 
mse = mean_squared_error(df['Y'], predicted_price_all)
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {mse**0.5:.2f}")
mae = mean_absolute_error(df['Y'], predicted_price_all) 
print(f"Mean Absolute Error: {mae:.2f}")
```
